### Implement K-Means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)
```

```python
pred = kmeans.predict(df)
print(pred)
```

```python
#Clusters
pd.Series(pred).value_counts()
```

### Evaluate a model

```python
kmeans.inertia_
```

```python
kmeans.score(df)
```

```python
#Create an empty list in order to store inertia values
SSE = []
for cluster in range(1, 21):
    kmeans = KMeans(n_jobs=-1, n_clusters=cluster)
    kmeans.fit(df)
    SSE.append(kmeans.inertia_)
```

```python
#Create a dataframe to store No of cluster and its inertia value
clusters_df = pd.DataFrame({'Cluster' : range(1, 21), 'SSE' : SSE})
clusters_df
```

```python
#Plot a graph of clusters vs inertia
plt.figure(figsize=(12, 6))
plt.plot(clusters_df['Cluster'], clusters_df['SSE'], marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
```

### Let's normalize the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled
```

```python
#Summary of a scaled data
pd.DataFrame(df_scaled).describe()
```

```python
#Create an empty list to store inertia value
SSE = []
for cluster in range(1, 21):
    kmeans = KMeans(n_jobs=-1, n_clusters=cluster)
    kmeans.fit(df_scaled)
    SSE.append(kmeans.inertia_)
```

```python
#Plot a graph of no of clusters vs its inertia
plt.figure(figsize=(12, 6))
frame_scaled = pd.DataFrame({'Cluster' : range(1, 21), 'SSE' : SSE})
plt.plot(frame_scaled['Cluster'], frame_scaled['SSE'], marker='o')
plt.xlabel('Cluster')
plt.ylabel('SSE')
```

```python
#Implement K-Means with K=4
kmeans = KMeans(n_clusters=4, n_jobs=-1)
kmeans.fit(df_scaled)
pred = kmeans.predict(df_scaled)
```

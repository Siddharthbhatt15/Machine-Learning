# Support Vector Classifier(SVC)

```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading a csv file
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
#Dimension of a dataframe
df.shape
```

```python
#First 10 rows of a dataframe
df.head(10)
```

```python
#Data types of all columns
df.dtypes
```

```python
#Checking for a missing values
df.isnull().sum()
```

```python
#Separating Independent and Dependent variable
X = df.iloc[:, [2, 3]].values
y = df.iloc[:, -1].values
```

```python
#Independent variable
print(X)
```

```python
#Dependent variable
print(y)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
print(X)
```

```python
#Splitting the dataset into Train and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11, stratify=y)
```

### Implement SVC(Support Vector Classifier)

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

```python
#Make a prediction for Test dataset
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
#Evaluate a model
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
print('Accuracy Score : ', accuracy_score(y_test, y_pred)*100)
print('Precision Score : ', precision_score(y_test, y_pred)*100)
print("Confusion Matrix : \n", confusion_matrix(y_test, y_pred))
```

### Visualizing the Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, svc.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

### Visualizing the Test set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, svc.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

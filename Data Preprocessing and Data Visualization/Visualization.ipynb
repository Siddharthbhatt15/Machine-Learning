### Understanding Percentiles

```python
#The quantile function
print(df['alcohol'].quantile(0.1))  # 0.1 or 10% values are below 9.1
print(df['alcohol'].quantile(0.25))  #0.25 or 25% values are below 9.5
print(df['alcohol'].quantile(0.5))    #0.5 or 50% values are below 10.3
print(df['alcohol'].quantile(0.75))    #0.75 or 75% values below 11.3
print(df['alcohol'].quantile(1.0))    #This is the maximum point
```

```python
df['alcohol'].min()
```

```python
df['alcohol'].max()
```

```python
#Quantile ranges
df['alcohol'].quantile([0.05, 0.95])
```

```python
df['alcohol'].quantile([0.10, 0.20,0.50, 0.80])
```

# Factor Plots and Bee Swarm Plots

Factor plots allow us to create separate plots by categorical classes.

```python
#Factor plot
g = sns.factorplot(x='free sulfur dioxide',
                  y='alcohol',
                   data=df,
                  # A recent update in Matplotlib broke how the hue argument works in Seaborn with some plots see: https://github.com/mwaskom/seaborn/issues/1697
                  #hue='quality',  # Color by stage
                  col='quality',
                   kind='point')    #Swarm point

#Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

# Creating Empirical cumulative distribution function (ECDF)

```python
import numpy as np
x = np.sort(df['alcohol'])
y = np.arange(1, len(x) + 1) / len(x)
_ = plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.05)
_ = plt.xlabel('Percent of alcohol in wine')
_ = plt.ylabel('ECDF')
plt.grid(True)
plt.show()
```

### Creating Bins

**Generating categorical variables from continuous variable by creating bins**

```python
bins = [0,12,19,30,60,90]
group=['child','teenager','young_adult','middle_aged', 'senior_citizen']
df['age_category'] = pd.cut(df['age'], bins, labels=group)
```

```python
df[['age', 'age_category']].head(10)
```

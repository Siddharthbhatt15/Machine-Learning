### Solution of Problem 1 : Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Outlet_Size'].value_counts()
```

```python
#Apply LabelEncoder
le.fit_transform(['Small', 'Medium', 'High'])
```

**LabelEncoder uses the alphabetical order**

```python
df['Outlet_Size'] = df['Outlet_Size'].map({'Small' : 0, 'Medium' : 1, 'High' : 2})
```

```python
df['Outlet_Size'].head()
```

So this solves the first challenge we encountered. Now we'll see how to deal with high cardinality.

### Solution of Problem 2

```python
df.nunique()
```

```python
df['Item_Identifier'].value_counts()
```

```python
temp = df['Item_Identifier'].value_counts()
temp.head()
```

```python
#Calculate a frequency of each item and store it into a new column
df['Item_Identifier_Count'] = df['Item_Identifier'].apply(lambda x : temp[x])
df[['Item_Identifier', 'Item_Identifier_Count']].head()
```

```python
for i in range(0, len(df)):
    if df['Item_Identifier_Count'][i] < 4:
        df['Item_Identifier'][i] = 'Other'
```

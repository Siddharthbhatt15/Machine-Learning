# Regression Benchmark

```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading a dataset
df = pd.read_csv('big_mart_sales.csv')
```

```python
#Dimension of dataset
df.shape
```

```python
#First 5 rows of the dataset
df.head()
```

```python
#Check for a missing values
df.isnull().sum()
```

```python
#Shuffling the dataset
from sklearn.utils import shuffle
df = shuffle(df, random_state=42)
```

```python
#View first 5 rows of dataset after shuffling
df.head()
```

```python
#Create 4 division of dataset
div = int(df.shape[0]/4)
div
```

```python
#Create train and test dataset
train = df.loc[:3*div+1]
test = df.loc[3*div+1:]
```

```python
#First 5 rows of train dataset
train.head()
```

```python
#First 5 rows of test dataset
test.head()
```

### Simple Mean(Mean of Item_Outlet_Sales)

```python
#Create a new column in test dataset named 'simple_mean' and store simple mean of Item_outlet_Sales columns of train data
test['Simple_Mean'] = train['Item_Outlet_Sales'].mean()
```

```python
#First 5 rows of test dataset
test.head()
```

```python
#Calculating Mean Absolute Error
from sklearn.metrics import mean_absolute_error as MAE
print("Mean Absolute Error : ", MAE(test['Item_Outlet_Sales'], test['Simple_Mean']))
```

### Mean of Item_Outlet_Sales with respect to Outlet_Type

```python
pd.pivot_table(train, values='Item_Outlet_Sales', index=['Outlet_Type'], aggfunc=np.mean)
```

```python
#Create a new column named 'Outlet_Type_Mean' and initialize to 0
test['Outlet_Type_Mean'] = 0
#For every unique entry in Outlet_Identifier
for i in train['Outlet_Type'].unique():
    #Assign a mean value corresponding to outlet_type
    test['Outlet_Type_Mean'][test['Outlet_Type'] == str(i)] = train['Item_Outlet_Sales'][train['Outlet_Type'] == str(i)].mean()
```

```python
#First 5 rows of test dataset
test.head()
```

```python
#Calculating Mean Absolute Error
print("Mean Absoluter Error : ", MAE(test['Item_Outlet_Sales'], test['Outlet_Type_Mean']))
```

### Mean of Item_Outlet_Sales with respect to Outlet_Establishment_Year

```python
pd.pivot_table(train, values ='Item_Outlet_Sales', index=['Outlet_Establishment_Year'], aggfunc=np.mean)
```

```python
#Create a new column and initialize it with 0
test['Outlet_Year_Mean'] = 0
#For every unique entry in Outlet_Identifier
for i in train['Outlet_Establishment_Year'].unique():
    #Assign a mean value corresponding to Outlet_Establishment_Year
    test['Outlet_Year_Mean'][test['Outlet_Establishment_Year'] == i] = train['Item_Outlet_Sales'][train['Outlet_Establishment_Year'] == i].mean()
```

```python
#First 5 rows of test dataset
test.head()
```

```python
#Calculating Mean Absolute Error
print("Mean Absolute Error : ", MAE(test['Item_Outlet_Sales'], test['Outlet_Year_Mean']))
```

### Mean of Item_Outlet_Sales with respect to Outlet_Location_Type

```python
pd.pivot_table(train, values='Item_Outlet_Sales', index='Outlet_Location_Type', aggfunc=np.mean)
```

```python
#Create a new column and initialize it to 0
test['Outlet_Location_Mean'] = 0
#For every unique entry in Item_Identifier
for i in train['Outlet_Location_Type'].unique():
    #Assign a mean value corresponding to a Outlet_Location_Type
    test['Outlet_Location_Mean'][test['Outlet_Location_Type'] == str(i)] = train['Item_Outlet_Sales'][train['Outlet_Location_Type'] == str(i)].mean()
```

```python
#First 5 rows of test dataset
test.head()
```

```python
#Calculating Mean Absolute Error
print("Mean Absolute Error : ", MAE(test['Item_Outlet_Sales'], test['Outlet_Location_Mean']))
```

### Mean of Item_Outlet_Sales with respect to both Outlet_Location_Type and Outlet_Establishment_Year

```python
pd.pivot_table(train, values='Item_Outlet_Sales', index=['Outlet_Location_Type', 'Outlet_Establishment_Year'], aggfunc=np.mean)
```

```python
#Create a new column and initialize it with 0
test['Super_Mean'] = 0

#Assigning variables to string(to shorten code length)
s1 = 'Outlet_Establishment_Year'
s2 = 'Outlet_Location_Type'

#For every unique value in s1
for i in test[s1].unique():
    #For every unique values in s2
    for j in test[s2].unique():
        #Calculate and assign mean to new column, corresponding to both unique values of s1 and s2 simultaneously
        test['Super_Mean'][(test[s1] == i) & (test[s2] == str(j))] = train['Item_Outlet_Sales'][(train[s1] == i) & (train[s2] == str(j))].mean()
```

```python
#First 5 rows of test dataset
test.head()
```

```python
#Calculate Mean Absolute Error
print("Mean Absolute Error : ", MAE(test['Item_Outlet_Sales'], test['Super_Mean']))
```
